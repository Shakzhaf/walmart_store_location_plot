## The Data Set

In the following cells, we will import our dataset from a .csv file as a Pandas DataFrame. Furthermore, we will begin exploring the dataset to gain an understanding of the type, quantity, and distribution of data in our dataset. For this purpose, we will use Pandas' built-in describe feature, as well as parameter histograms and a correlation matrix.

```python
# Load the dataset from the csv file using pandas
df=pd.read_csv('/content/walmart_store_locaction.csv.zip')
```

```python
# Drop all columns which have nulll data
df.drop(columns = ['phone_number_2','fax_2','fax_1','email_1','email_2','website','facebook','instagram','pinterest','youtube','twitter'], inplace = True)
```

```python
# Plot all walmart present in USA on a scatterplot
cmap = sns.color_palette(as_cmap=True)
sns.set(rc={'figure.figsize':(11,8)})
sns.scatterplot(data=df[df['country']=='US'], x="longitude", y="latitude", palette = cmap,  
                sizes=(20, 200), color = ['teal'])
plt.title('Walmart Stores Location')
plt.xlabel('longitude')
plt.ylabel('latitude')
plt.show()
```

```python
# Plot all walmart present in Texas(USA) on a scatterplot
cmap = sns.color_palette(as_cmap=True)
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.scatterplot(data=df[df['state']=='TX'], x="longitude", y="latitude", palette = cmap,  
                sizes=(20, 200), color = ['teal'])
plt.title('Walmart Stores Location in TX')
plt.xlabel('longitude')
plt.ylabel('latitude')

plt.show()
```

```python
# Plot all walmart present in San Antonio, Texas, USA on a scatterplot
cmap = sns.color_palette(as_cmap=True)
sns.set(rc={'figure.figsize':(11,8)})
sns.scatterplot(data=df[df['city']=='San Antonio'], x="longitude", y="latitude", palette = cmap,  
                sizes=(20, 200), color = ['teal'])
plt.title('Walmart Stores Location in San Antonio')
plt.xlabel('longitude')
plt.ylabel('latitude')

plt.show()
```
